# Series 연산

```python
s = pd.Series([9904312, 3448737, 2890451, 2466052],
             index=["서울", "부산", "인천", "대구"])
s2 = pd.Series({
    "서울" : 9631482,
    "부산" : 3393191,
    "인천" : 2632035,
    "대전" : 1490158
}, index=["부산", "서울", "인천", "대전"])
ds = s - s2
print(ds)
```

```python
s3 = s / 1000000
print(s3)
```

```python
print(s.values, s2.values)
print(s.values - s2.values)
```

```python
rs = (s - s2) / s2 * 100
print(rs)
```

```python
rs = rs[rs.notnull()]
print(rs)
```

```python
rs["부산"] = 1.63
rs["대구"] = 1.41
print(rs)
```

```python
del rs["서울"]
print(rs)
```

```python
def remove_comma(x):
    return int(x.replace(",", ""))
```

```python
s = Series(["1,234", "5,678", "9,876"])
result = list(s.map(remove_comma))
print(result)
```
