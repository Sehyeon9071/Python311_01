```python
data = [ ["2차전지(생산)", "SK이노베이션", 10.19, 1.29], ["해운", "팬오션", 21.23, 0.95], ["시스템반도체", "티엘아이", 35.97, 1.12], ["해운", "HMM", 21.52, 3.20], ["시스템반도체", "아이에이", 37.32, 3.55], ["2차전지(생산)", "LG화학", 83.06, 3.75] ]

columns = ["테마", "종목명", "PER", "PBR"]

df = pd.DataFrame(data=data, columns=columns)
```

```python
df
```

(1) 데이터프레임에서 테마별 PER, PBR의 평균을 계산하세요.

```python
df3 = df.groupby("테마").mean(numeric_only=True)
df3
```

(2) 데이터프레임에서 테마별 종목의 개수를 계산하세요.

```python
df3 = df.groupby("테마").agg({"종목명":"count"})
df3
```

(3) 데이터프레임에서 테마별 PER의 평균을 계산하세요.

```python
df3 = df.groupby("테마").agg({"PER":"mean"})
df3
```

(4) 데이터프레임에 대해 테마로 그룹화한 후 PER, PBR 컬럼을 슬라이싱하세요. 그 다음 두 컬럼에 대한 평균을 계산하세요.

```python
df3 = df.groupby("테마")[["PER","PBR"]].mean()
df3
```

(5) 데이터프레임에 대해서 '테마' 컬럼으로 그룹화한 후 PER 컬럼은 max, PBR 컬럼은 min 함수를 적용해보세요.

```python
df3 = df.groupby("테마").agg({"PER":"max","PBR":"min"})
df3
```

# [문제3]

아래 코드를 활용해서 주어진 문제에 대한 코드를 작성하시오.

import pandas as pd

data1 = {
    'ID': [100, 101, 102, 103],
    'Name': ['Brayden', 'Clara', 'Ella', 'Oliver']
}

data2 = [
    [100, "삼성전자"],
    [100, "LG전자"],
    [101, "카카오"],
    [101, "NAVER"],
    [101, "현대차"]
]

df1 = pd.DataFrame(data1)

df2 = pd.DataFrame(data=data2, columns=['ID', '종목명'])


```python
import pandas as pd

data1 = {
    'ID': [100, 101, 102, 103],
    'Name': ['Brayden', 'Clara', 'Ella', 'Oliver']
}

data2 = [
    [100, "삼성전자"],
    [100, "LG전자"],
    [101, "카카오"],
    [101, "NAVER"],
    [101, "현대차"]
]

df1 = pd.DataFrame(data1)

df2 = pd.DataFrame(data=data2, columns=['ID', '종목명'])
```

```python
df1
```

```python
df2
```

(1) df1과 df2를 'ID'라는 컬럼명을 기준으로 'inner' 방식으로 병합하세요.

```python
df3 = pd.merge(df1, df2, on="ID", how="inner")
df3
```

(2) df1과 df2를 'ID’라는 컬럼명을 기준으로 ‘outer’ 방식으로 병합하세요.

```python
df3 = pd.merge(df1, df2, on="ID", how="outer")
df3
```

(3) 데이터프레임에서 df1의 'ID'와 df2의 '아이디' 컬럼을 기준으로 merge 하세요.

```python
df2 = pd.DataFrame(data=data2, columns=['아이디', '종목명'])
df3 = pd.merge(df1, df2, left_on="ID", right_on="아이디")
df3
```

import pandas as pd

```python
import pandas as pd

data1 = {
    'ID': [100, 101, 102, 103],
    'Name': ['Brayden', 'Clara', 'Ella', 'Oliver']
}

data2 = [
    [100, "삼성전자"],
    [100, "LG전자"],
    [101, "카카오"],
    [101, "NAVER"],
    [101, "현대차"]
]

df1 = pd.DataFrame(data1)

df1.set_index('ID', inplace=True)

df2 = pd.DataFrame(data=data2, columns=['ID', '종목명'])

df2.set_index('ID', inplace=True)
```

```python
df1
```

```python
df2
```

```python
df3 = pd.merge(df1, df2, on="ID")
df3
```
