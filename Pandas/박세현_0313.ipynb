(1) iris 데이터에서 species별 sepal_length의 평균을 구한 후 이를 'bar' 타입으로 그려보세요.

(2) iris 데이터에서 품좀별 데이터의 개수를 파이 차트 형태로 그려보세요.

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
```

```python
iris
```

```python
bar_iris = iris.groupby("species")["sepal_length"].mean()
bar_iris.plot.bar(rot=0)
plt.show()
```

```python
pie_iris = iris["species"].value_counts()
pie_iris.plot.pie(autopct="%.2f%%")
plt.show()
```
